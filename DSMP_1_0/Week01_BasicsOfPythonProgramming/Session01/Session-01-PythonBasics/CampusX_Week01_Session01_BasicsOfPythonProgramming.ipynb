## 3. **Variables**

```python
# Static Vs Dynamic Typing
# Static Vs Dynamic Binding
# Stylish Declaration Techniques

name = 'Munish' # no need to show data type....
print (name)

a = 5
b = 6
print (a + b)

name1 = 'Munish'
blankspace = ' '
name2 = 'Sawhney'
print (name1 + blankspace + name2)

# '+' used to add for two numbers
# '+' used to concatenate two strings / text
```

```python
# Dynamic Typing
# you do not tell the type of variable. Python automatically infers type
# you need to tell the type of variable like C/ C++ / Java

# Dynamic binding -> supported by python
a = 5
print (a, type(a))
a = 'Munish'
print (a, type(a))

# Static binding -> supported by c / c++ / java
# cannot change the type once assigned
```

```python
# Dynamic binding -> supported by python

a = 1
b = 2
c = 3
print(a,b,c)

```

```python
# Stylish Declaration Techniques
a,b,c = 1,2,3
print(a,b,c)
print(a,b)
print(b,a)

same_value_var1=same_value_var2=same_value_var3 = 5
print(same_value_var1, same_value_var2, same_value_var3)
```

# 6. **Type Conversion**

```python
# in type conversion the variable is not permanently converted but in typecasting the varialbe is permanently converted.
# Implicit type conversion vs Explicit type conversion
# Implicit type conversion = compiler automatically converts the type
# eg....
print (5 + 5.6)
print (type(5), type(5.6), type(5 + 5.6))
# 5 is implicitly converted to float

# eg. print(4 + '4') -> gives error as int and str are incompatible

# explict type conversion
# str -> int
# type conversion is possible if it is possible
print(type(int('4')))
print(type(complex('4+5j')))
print(type(str(5)))
```

```python
fnum = input("Enter the 1st number: ")
snum = input("Enter the 2nd number: ")
print(f"fnum: {fnum} and type(fnum): {type(fnum)}")
print(f"snum: {snum} and type(snum): {type(snum)}")
result = int(fnum) + int(snum)
print(f"result: {result}")
print(f"fnum type: {type(fnum)}")
```

# 7. **Literals**





```python
# What ever raw value is stored is called as Literal.

a = 0b1010 # Binary Literal
b = 100  # Decimal Literal
c = 0o310 # Octal Literal
d = 0x12c # Hexadecimal Literal
print(a, b, c, d)

# Float Literal
float_1 = 10.5
float_2 = 1.5e2 # 1.5 * 10^2
float_3 = 1.5e-3  # 1.5 * 10^-3

# Complex Literal
x = 3.14j
print(x.real, x.imag)

```

```python
# string literals
string = 'This is Python'
strings = "This is Python"
char = "C"
multiline_str = """This is a multiline string with more than one line code."""
unicode = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(c)
print(multiline_str)
print(unicode)
print(raw_str)
```
